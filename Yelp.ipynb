```python
# calculating the distance between two points (in m) using the Haversine formula

# approximate radius of earth in km


def distance_meters (lat1, lon1, lat2, lon2):

    R = 6373.0

    lat1_radian = radians(lat1)
    lon1_radian = radians(lon1)
    lat2_radian = radians(lat2)
    lon2_radian = radians(lon2)

    dlon = lon2_radian - lon1_radian
    dlat = lat2_radian - lat1_radian

    a = sin(dlat / 2)**2 + cos(lat1_radian) * cos(lat2_radian) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance_km = R * c
    distance = distance_km/1000

    return distance

distance_meters(52.36575, 4.94142, 52.366211, 4.859380)
```
